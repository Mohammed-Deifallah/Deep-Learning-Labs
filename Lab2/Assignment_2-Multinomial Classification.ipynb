# Assignment 2-Multinomial Classification


## Problem Statement
In this assignment we will be addressing the task of multinominal classification of handwrittendigits from the famous `MNIST` dataset.The MNIST dataset consists of $60,000$ training images and $10,000$ test images. Our classes arethe digits $0$-$9$.You are required to build $2$ models to solve this problem:
1.  A simple using fully connected layers.
2. A model using Convolutional Neural Network (CNN) before applying the simple model.

You should observe the difference in performance between the 2 models.

## Imports

```python
import tensorflow as tf
tf.enable_eager_execution()
import matplotlib.pyplot as plt
import numpy as np
import random
from progressbar import progressbar
```

```python
from sklearn.model_selection import train_test_split
```

```python
%config InlineBackend.figure_format = 'retina' #set 'png' here when working on notebook
%matplotlib inline
```

## Data Reading and Preprocessing

### Reading

```python
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

### Preprocessing

```python
train_images = np.expand_dims(train_images, axis=-1)/255.
train_labels = np.int64(train_labels)
train_labels = tf.keras.utils.to_categorical(train_labels, num_classes=10)
test_images = np.expand_dims(test_images, axis=-1)/255.
test_labels = np.int64(test_labels)
test_labels = tf.keras.utils.to_categorical(test_labels, num_classes=10)
train_images, val_images, train_labels, val_labels = train_test_split(train_images, train_labels, test_size=0.33, random_state=42)
```

### Samples Visualization

```python
plt.figure(figsize=(10,10))
random_inds = np.random.choice(len(train_images),36)
for i in range(36):
    plt.subplot(6,6,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    image_ind = random_inds[i]
    plt.imshow(np.squeeze(train_images[image_ind]), cmap=plt.cm.binary)
    #plt.xlabel(train_labels[image_ind][train_labels[image_ind] == 1])
```

```python
def plot(history, accuracy, optimizer, lr=.01):
    plt.plot(history.history['acc'])
    plt.title('Optimizer: ' + str(optimizer) + ', Learning_rate: ' + str(lr) + ', Accuracy: ' + str(accuracy*100) + '%')
    plt.xlabel('epoch')
    plt.legend(loc='upper left')
    plt.show()
```

## Model

### FC Model


```python
def build_fc_model():
    fc_model = tf.keras.Sequential([
      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(128, input_dim=784, activation='relu'),
      tf.keras.layers.Dense(10, activation='softmax')
    ])
    return fc_model
```

### CNN Model


```python
def build_cnn_model():
    cnn_model = tf.keras.Sequential([
        tf.keras.layers.Conv2D(filters=24, kernel_size=(3,3), input_shape=(28, 28, 1), activation=tf.nn.relu),       
        tf.keras.layers.MaxPool2D(pool_size=(2,2)),
        tf.keras.layers.Conv2D(filters=36, kernel_size=(3,3), activation=tf.nn.relu),
        tf.keras.layers.MaxPool2D(pool_size=(2,2)),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, input_dim=784, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
    ])
    return cnn_model
```

## Hyperparameter Tuning

### FC Model

```python
model = build_fc_model()
BATCH_SIZE = 64
EPOCHS = 5
best_lr=1
```

```python
optimizers = [tf.train.AdagradOptimizer, tf.train.AdadeltaOptimizer, tf.train.AdamOptimizer,
              tf.train.GradientDescentOptimizer, tf.train.RMSPropOptimizer]
lrs = [.001, .01, .05, .1, 1, 10, 100]
```

```python
max_acc = -1
best_opt_fc = tf.train.AdagradOptimizer(best_lr)
```

```python
for optimizer in optimizers:
    model = build_fc_model()
    model.compile(optimizer=optimizer(best_lr), loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(train_images, train_labels, verbose = 0, epochs=EPOCHS, batch_size=BATCH_SIZE)
    score = model.evaluate(val_images, val_labels, batch_size=BATCH_SIZE)
    print(score)
    plot(history, score[1], optimizer, best_lr)
    if score[1] > max_acc:
        max_acc = score[1]
        best_opt_fc = optimizer
```

```python
max_acc=-1
best_lr_fc=1
```

```python
for lr in lrs:
    model = build_fc_model()
    model.compile(optimizer=best_opt_fc(lr), loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(train_images, train_labels, verbose = 0, epochs=EPOCHS, batch_size=BATCH_SIZE)
    score = model.evaluate(val_images, val_labels, batch_size=BATCH_SIZE)
    plot(history, score[1], best_opt_fc, lr)
    if score[1] > max_acc:
        max_acc = score[1]
        best_lr_fc = lr
```

## Evaluation

```python
from time import time
```

```python
fc_model = build_fc_model()
start = time()
fc_model.compile(optimizer=best_opt_fc(best_lr_fc), loss='categorical_crossentropy', metrics=['accuracy'])
history = fc_model.fit(train_images, train_labels, verbose = 0, epochs=EPOCHS, batch_size=BATCH_SIZE)
end = time()
score = fc_model.evaluate(test_images, test_labels, batch_size=BATCH_SIZE)
plot(history, score[1], best_opt_fc, best_lr_fc)
print("Time: " + str(end - start))
```
